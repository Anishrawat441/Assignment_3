# Task 2 - Recursive Function: Factorial Utility

### Create a recursive function factorial(n) that:

```python
# 1. Returns the factorial of n.
def factorial(n):
    if n == 0 or n == 1:   # base case
        return 1
    else:
        return n * factorial(n - 1)
```

```python
print(factorial(5))
```

```python
# 2. Handles two edges cases:n == 0 and n ==1.
def factorial(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
print(factorial(0))
print(factorial(1))
```

```python
# 3.prints an error message if n is negative.
def factorial(n):
    if n < 0:
        print("Error: Factorial is not defined for negative numbers.")
        return None
    elif n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return n * factorial(n - 1)
```

```python
print(factorial(-8))
```

```python
print(factorial(5))
```
