# Task 5- Using filter(): Filter Expensive Products

### Given the list:
### Prices=[100, 250 ,400 ,1200 ,50, 2000, 850]
### Use filter() to :

```python
# 1. Create a list of prices greater then 500.
Prices = [100, 250, 400, 1200, 50, 2000, 850]

# Filter prices >500
high_prices = list(filter(lambda price: price > 500, Prices))

print(high_prices)
```

```python
# 2. Create another list of prices less then or equal to 500.

Prices = [100, 250, 400, 1200, 50, 2000, 850]

# Filter prices <=500
low_prices = list(filter(lambda price: price <= 500, Prices))

print(low_prices)
```
